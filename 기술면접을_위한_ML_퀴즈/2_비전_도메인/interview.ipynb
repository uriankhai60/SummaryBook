# Q-001) CNN과 Vision Transformer의 구조적 차이와 장단점은 무엇입니까?

## 답변  
이미지 특징을 추출하는 방법에서 근본적인 차이점이 존재합니다.  
CNN의 경우 커널의 사이즈로 한정된 이미지의 지역적인 특징을 추출합니다.  
layer의 깊이가 깊어질수록 지역적인 특징(모서리, 패턴)에서 전역적인 특징 추출로 발전합니다.  
  
ViT의 경우 Patchify를 통해서 각 지역을 토큰으로 처리하고 Self-Attention을 이용.  
모든 패치간의 상호 관계를 학습합니다.  
처음부터 전역적인 피쳐를 추출합니다.  
  
CNN의 경우 파라미터 효율적이며, 적은 데이터로도 학습이 용이합니다.  
또한 국소적인 패턴인식에 유리한 부분이 있으나, 이미지의 전체적인 특징을 잡아내는것에는 불리합니다.  

Transformer의 경우 이미지의 전체적인 특징을 잘 잡아내며 전이학습에 유리합니다.  
하지만, 데이터와 메모리가 많이 필요하며, From scratch 학습시 많은 추가적인 테크닉을 필요로 합니다.  
  
사용하는 도메인에 따라서 전역적인 특징이 중요한 경우와 지역적인 특징이 중요한 경우  
보유한 데이터의 양 등을 고려하여 피쳐 익스트렉터를 잘 사용하는 것이 중요합니다.
