# Q-001) CNN과 Vision Transformer의 구조적 차이와 장단점은 무엇입니까?

## 답변  
이미지 특징을 추출하는 방법에서 근본적인 차이점이 존재합니다.  
CNN의 경우 커널의 사이즈로 한정된 이미지의 지역적인 특징을 추출합니다.  
layer의 깊이가 깊어질수록 지역적인 특징(모서리, 패턴)에서 전역적인 특징 추출로 발전합니다.  
  
ViT의 경우 Patchify를 통해서 각 지역을 토큰으로 처리하고 Self-Attention을 이용.  
모든 패치간의 상호 관계를 학습합니다.  
처음부터 전역적인 피쳐를 추출합니다.  
  
CNN의 경우 파라미터 효율적이며, 적은 데이터로도 학습이 용이합니다.  
또한 국소적인 패턴인식에 유리한 부분이 있으나, 이미지의 전체적인 특징을 잡아내는것에는 불리합니다.  

Transformer의 경우 이미지의 전체적인 특징을 잘 잡아내며 전이학습에 유리합니다.  
하지만, 데이터와 메모리가 많이 필요하며, From scratch 학습시 많은 추가적인 테크닉을 필요로 합니다.  
  
사용하는 도메인에 따라서 전역적인 특징이 중요한 경우와 지역적인 특징이 중요한 경우  
보유한 데이터의 양 등을 고려하여 피쳐 익스트렉터를 잘 사용하는 것이 중요합니다.

# Q-002) Semantic/Instance/Panoptic Segmentation의 차이는 무엇인가?

## 답변  
Semantic Segmentation은 가장 기초적인 Seg 모델로써 각 픽셀을 클래스로 분류하는 모델입니다.  
모델의 output은 classification과 유사하게 각 픽셀이 사전의 정의된 크래스중 하나의 클래스  
(logit->argmax)로 반드시 할당됩니다.  
다만 동일 클래스 내의 개별 객체 구분이 불가능합니다.  (e.g.) 개체간의 사람1, 사람2간 구분이 되지 않음  

Instance Segmentation은 동일 클래스 내에서도 객체 단위로 구분합니다.  
일반적으로 Object Detection을 통해서 ROI 영역을 예측한 뒤에 픽셀단위 마스크를 예측하는 구조입니다.  
장점으로는 각 개체간 인스턴스 구분을 할 수 있으나.  
단점으로는 올바른 RoI가 예측되었다는 가정하에 Segmentation이 수행되기 때문에 추론시간이 늦고 모델이 다소 복잡한 편입니다.  
또한 한 픽셀이 두가지 Instance ID에 할당되는듯 중복 문제도 발생할 수 있습니다.  
  
Panoptic Segmentation은 두가지 모델의 단점을 개선한 모델입니다.  
한 픽셀에 라벨이 class와 instance를 모두 추론하며 ROI과정이 없기 때문에 배경(~=stuff)영역의 구분도 가능합니다.  
ROI에서 마스크를 예측하는 Instance Segmentation과 달리 전체 이미지를 end2end로 수행하기 때문에  
중복이 나타나지 않습니다.
